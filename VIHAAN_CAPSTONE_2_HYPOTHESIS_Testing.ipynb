**Does Existing Defensive Team have any significance with any other field in the dataset.**

**Chi-Squared Test** (category vs category)

```python
for i in catcol:
  print(i)
  contingency_table=pd.crosstab(data['DefensiveTeam'],data[i])
  print('contingency_table :-\n',contingency_table)
#Observed Values
  Observed_Values = contingency_table.values 
  print("Observed Values :-\n",Observed_Values)
  b=stats.chi2_contingency(contingency_table)
  Expected_Values = b[3]
  print("Expected Values :-\n",Expected_Values)
  no_of_rows=len(contingency_table.iloc[0:2,0])
  no_of_columns=len(contingency_table.iloc[0,0:2])
  ddof=(no_of_rows-1)*(no_of_columns-1)
  print("Degree of Freedom:-",ddof)
  alpha = 0.05
  from scipy.stats import chi2
  chi_square=sum([(o-e)**2./e for o,e in zip(Observed_Values,Expected_Values)])
  chi_square_statistic=chi_square[0]+chi_square[1]
  print("chi-square statistic:-",chi_square_statistic)
  critical_value=chi2.ppf(q=1-alpha,df=ddof)
  print('critical_value:',critical_value)
#p-value
  p_value=1-chi2.cdf(x=chi_square_statistic,df=ddof)
  print('p-value:',p_value)
  print('Significance level: ',alpha)
  print('Degree of Freedom: ',ddof)
  print('chi-square statistic:',chi_square_statistic)
  print('critical_value:',critical_value)
  print('p-value:',p_value)
  if chi_square_statistic>=critical_value:
    print("Reject H0,There is a relationship between 2 categorical variables")
  else:
    print("Retain H0,There is no relationship between 2 categorical variables")
    
  if p_value<=alpha:
    print("Reject H0,There is a relationship between 2 categorical variables")
  else:print("Retain H0,There is no relationship between 2 categorical variables")
  print('*'*50)
```

**Annova test**  (category vs numerical)

```python
for i in arr:
  print(i)
  df_anova = data[['DefensiveTeam',i]]
  grps = pd.unique(df_anova['DefensiveTeam'].values)
  d_data = {grp:df_anova[i][df_anova['DefensiveTeam'] == grp] for grp in grps}
 
  F, p = stats.f_oneway(d_data['PIT'], d_data['TEN'], d_data['MIN'], d_data['DET'], d_data['DAL'],d_data['PIT'], d_data['TEN'], d_data['MIN'], d_data['DET'], d_data['DAL'],
                        d_data['HOU'], d_data['NYJ'], d_data['JAC'], d_data['DEN'], d_data['PHI'],d_data['CAR'], d_data['KC'], d_data['BAL'], d_data['ATL'], d_data['ARI'],
                        d_data['SF'], d_data['SEA'], d_data['STL'], d_data['WAS'], d_data['NYG'], d_data['NE'], d_data['OAK'], d_data['SD'], d_data['CIN'], d_data['GB'],
                         d_data['BUF'], d_data['TB'], d_data['CLE'], d_data['CHI'], d_data['IND'], d_data['MIA'], d_data['LA'], d_data['JAX'])
  print("p-value for significance is: ", p)
  if p<0.05:
    print("reject null hypothesis")
    print('*'*50)
  else:
    print("accept null hypothesis")
    print('*'*50)
```
